```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
import sys
in_colab = 'google.colab' in sys.modules

if in_colab:
  BASE_DIR = "https://github.com/DiploDatos/AnalisisYCuracion/raw/master/"
else:
  BASE_DIR = ".."
```

```python
tabla_1 = pd.read_csv(BASE_DIR + "/input/wtageinf.csv")
tabla_1.head(10)
```

```python
filtrado= tabla_1[tabla_1.Sex==1]

#filtrado.plot(x='Agemos', y='P3')

plt.style.use('bmh')
plt.plot(filtrado.Agemos, filtrado.P97, label='P97')
plt.plot(filtrado.Agemos, filtrado.P95, label='P95')
plt.plot(filtrado.Agemos, filtrado.P90, label='P90')              
plt.plot(filtrado.Agemos, filtrado.P75, label='P75')
plt.plot(filtrado.Agemos, filtrado.P50, label='P50')
plt.plot(filtrado.Agemos, filtrado.P25, label='P25')
plt.plot(filtrado.Agemos, filtrado.P10, label='P10')
plt.plot(filtrado.Agemos, filtrado.P5, label='P5')
plt.plot(filtrado.Agemos, filtrado.P3, label='P3')
plt.title("Percentiles de peso de nios")
plt.ylabel("Peso")
plt.xlabel("Edad en meses")
plt.legend()
plt.show()

#filtrado.plot('Agemos', 'P5');
```
